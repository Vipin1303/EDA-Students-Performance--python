```python
import cv2
import os

os.makedirs("frames", exist_ok=True)

video_path = "ecwaxgutkc.mp4"  # Just the name, not full path
cap = cv2.VideoCapture(video_path)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    filename = f"frames/frame_{frame_count:04d}.jpg"
    cv2.imwrite(filename, frame)
    frame_count += 1

cap.release()
print("Total frames extracted:", frame_count)
```

```python
from mtcnn import MTCNN
import cv2
import os
from PIL import Image

detector = MTCNN()


#folder
input_folder = "frames"
output_folder = "faces"
os.makedirs(output_folder, exist_ok=True)


for image_file in os.listdir(input_folder):
    path = os.path.join(input_folder, image_file)


    image = cv2.imread(path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)   # Convert to RGB (because MTCNN works with RGB)
    faces = detector.detect_faces(image_rgb)


    for i, face in enumerate(faces):
        x, y, width, height = face['box']
        x, y = max(0, x), max(0, y)


        face_image = image_rgb[y:y+height, x:x+width]


        save_name = f"{output_folder}/{image_file[:-4]}_face{i}.jpg"
        Image.fromarray(face_image).save(save_name)

print("All faces saved in 'faces' folder.")
```
